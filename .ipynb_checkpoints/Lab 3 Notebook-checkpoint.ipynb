The goal of this lab is to plot a histogram of the words in the play "Romeo and Juliet".

We begin by reading the words in the text and creating a dictionary to store the frequency counts for unique words in a dictionary `words`.

```python
%matplotlib inline
import matplotlib.pyplot as plt

file = open("romeoandjuliet.txt", "r")
words = {}
for line in file:
    for word in line.split():
        word = word.lower()
        word = ''.join(filter(str.isalpha, word))
        if word not in words:
            words[word] = 1
        else:
            words[word] += 1
```

Next, we sort the words according to their frequencies and then extract the frequencies, storing them in a list `frequencies` to be plotted along with word ranks.

```python
sorted_words = sorted(words.items(), key=lambda kv: kv[1], reverse=True)
frequencies = [wf[1] for wf in sorted_words]
```

Now, we plot the frequencies of each word with the word rank (the $i^{th}$ rank corresponds to the $i^{th}$ most common word).

```python
plt.figure(figsize=(50, 50))
plt.bar([i + 1 for i in range(len(frequencies))], frequencies)
plt.xlabel("Word rank")
plt.ylabel("Frequency by count")

plt.show()
```

The code snippets below plot the 10 most common and 10 least common words, respectively.

```python
ten_most_frequent = sorted_words[0:10]
plt.bar([wf[0] for wf in ten_most_frequent], [wf[1] for wf in ten_most_frequent])
```

```python
ten_least_frequent = sorted_words[-11: -1]
plt.figure(figsize=(10, 5))
plt.bar([wf[0] for wf in ten_least_frequent], [wf[1] for wf in ten_least_frequent])
```

The ten most common words are `and`, `the`, `i`, `to`, `a`, `of`, `my`, `that`, `is`, `in`. These words are quite short in length; they are generally conjunctions, prepositions, articles, and simple verbs or pronouns. This may suggest that the English language (or "Romeo and Juliet" at the very least) relies heavily on these parts of speech for communication. They could be essential for conveying the most basic of ideas.

The ten least common words are `punishd`, `jointure`, `demand`, `statue`, `whiles`, `figure`, `sacrifices`, `glooming`, `pardond`, and `punished`. In contrast to the most common words, these words are longer in length; they are generally specific verbs and adjectives or adverbs. This may suggest that these parts of speech are non-essential; in fact, there may be other synonyms that can be used in place to communicate similar ideas. Additionally, these words may be less common simply because the context of "Romeo and Juliet" utilizes them less frequently.

```python
plt.figure(figsize=(10, 10))
plt.loglog([i + 1 for i in range(len(frequencies))], frequencies)
plt.xlabel("Word rank")
plt.ylabel("Frequency by count")

plt.show()
```

The log-log plot of frequency vs. word rank roughly looks linear. This seems to suggest that the distribution of frequencies and their associated words obeys a power law (short words are common, while longer words are less common). That is, the frequency of a word is approximately inversely proportional to its rank. This is known as **Zipf's Law**. Let $f(r)$ be the frequency of a word with rank $r$. Then, for $\alpha \approx 1$:

$$f(r) \propto \frac{1}{r^\alpha}$$

`Piantadosi, S.T. Psychon Bull Rev (2014) 21: 1112. https://doi.org/10.3758/s13423-014-0585-6`
