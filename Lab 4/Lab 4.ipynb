The code below generates a SQLite3 database `lab4.db`, populaing it with the three tables. After, we loop through the nested commands in `all_inserts`, inserting entries into each table using `c.execute(command)`.
```python
conn = sqlite3.connect('lab4.db')
c = conn.cursor()
c.execute("""CREATE TABLE genes (name TEXT, 
                                 description TEXT, 
                                 organism TEXT,
                                 nucleotide_sequence TEXT);""")
c.execute("""CREATE TABLE pathways (name TEXT, 
                                    description TEXT); """)
c.execute("""CREATE TABLE enzymes (name TEXT,
                                   function TEXT,
                                   ec TEXT);""")
all_inserts = [gene_table_inserts, pathway_table_inserts, enzyme_table_inserts]
for table_insert in all_inserts:
    for command in table_insert:
        c.execute(command)
```
